```python
import random
import csv
import base64

# Simulación de base de usuarios
roles = {
    "dueno": {"lectura": True, "escritura": True, "desencriptar": True, "añadir": True, "eliminar": True},
    "entrenador": {"lectura": True, "escritura": True, "desencriptar": "parcial", "añadir": True, "eliminar": False},
    "nutricionista": {"lectura": True, "escritura": False, "desencriptar": "imc", "añadir": False, "eliminar": False},
    "recepcionista": {"lectura": True, "escritura": True, "desencriptar": False, "añadir": True, "eliminar": False},
    "miembro": {"lectura": True, "escritura": False, "desencriptar": False, "añadir": False, "eliminar": False}
}

# Funciones de encriptación
def encriptar(texto):
    return base64.b64encode(texto.encode()).decode()

def desencriptar(texto_encriptado):
    return base64.b64decode(texto_encriptado.encode()).decode()

# Datos base
nombres = ["Carlos", "María", "Juan", "Andrea", "Luis", "Sofía", "Pedro", "Camila", "Javier", "Paula"]
apellidos = ["Gómez", "Rodríguez", "Martínez", "Fernández", "Pérez", "López", "Torres", "Ramírez", "Vargas", "Morales"]
razones = ["Bajar de peso", "Ganar músculo", "Rehabilitación física", "Reducir estrés", "Aumentar energía"]

# Crear CSV protegido
archivo_csv = "personas_roles.csv"
with open(archivo_csv, mode='w', newline='') as archivo:
    escritor = csv.writer(archivo)
    escritor.writerow([
        "ID (enc)", "Nombre y Apellido (enc)", "Número", "Motivo", "Peso", "Altura", "BPM", "IMC", "Clasificación"
    ])
    for _ in range(50):
        nombre = f"{random.choice(nombres)} {random.choice(apellidos)}"
        peso = round(random.uniform(50, 100), 1)
        altura = round(random.uniform(1.5, 2.0), 2)
        imc = round(peso / altura**2, 1)
        clasif = ("Bajo peso" if imc < 18.5 else
                  "Normal" if imc < 25 else
                  "Sobrepeso" if imc < 30 else "Obesidad")
        escritor.writerow([
            encriptar(str(random.randint(10000000, 99999999))),
            encriptar(nombre),
            f"3{random.randint(100000000, 999999999)}",
            random.choice(razones),
            peso, altura,
            random.randint(60, 120),
            imc, clasif
        ])

print("✅ Archivo generado con roles de acceso.\n")

# === ACCESO SIMULADO ===
def acceder_archivo_como(rol_usuario):
    permisos = roles.get(rol_usuario, {})
    print(f"🔑 Accediendo como '{rol_usuario}'")

    if not permisos.get("lectura", False):
        print("⛔ Acceso denegado para leer los datos.")
        return

    with open(archivo_csv, mode='r') as archivo:
        lector = csv.reader(archivo)
        encabezado = next(lector)
        print("Campos:", encabezado)
        for i, fila in enumerate(lector):
            if i >= 3:
                break
            id_enc = fila[0]
            nombre_enc = fila[1]
            imc = fila[7]
            clasificacion = fila[8]

            if permisos.get("desencriptar") == True:
                id_real = desencriptar(id_enc)
                nombre_real = desencriptar(nombre_enc)
            elif permisos.get("desencriptar") == "parcial":
                id_real = "[PROTEGIDO]"
                nombre_real = desencriptar(nombre_enc)
            elif permisos.get("desencriptar") == "imc":
                id_real = "[PROTEGIDO]"
                nombre_real = "[PROTEGIDO]"
            else:
                id_real = "[PROTEGIDO]"
                nombre_real = "[PROTEGIDO]"

            print(f"{i+1}. ID: {id_real}, Nombre: {nombre_real}, IMC: {imc}, Estado: {clasificacion}")

# Simular acceso
acceder_archivo_como("dueno")
print("\n---\n")
acceder_archivo_como("entrenador")
print("\n---\n")
acceder_archivo_como("miembro")
```

```python
# ⚙️ Configuración de roles y niveles
roles = {
    "miembro": {
        "nivel": 0,
        "acciones": ["ver datos"]
    },
    "recepcionista": {
        "nivel": 1,
        "acciones": ["ver datos", "añadir persona"]
    },
    "entrenador": {
        "nivel": 2,
        "acciones": ["ver datos", "añadir persona", "modificar persona", "ver IMC"]
    },
    "dueno": {
        "nivel": 4,
        "acciones": ["ver datos", "añadir persona", "modificar persona", "eliminar persona", "ver IMC", "modificar roles"]
    }
}

# Simulador de base de datos
personas = [
    {"nombre": "Carlos G.", "peso": 70, "altura": 1.75},
    {"nombre": "María F.", "peso": 58, "altura": 1.62}
]

# Mostrar menú de acciones válidas
def mostrar_menu(rol):
    acciones = roles[rol]["acciones"]
    print(f"\n🎮 Acciones disponibles para el rol '{rol}':")
    for i, accion in enumerate(acciones, 1):
        print(f"{i}. {accion}")
    return acciones

# Ejecutar acción seleccionada
def ejecutar_accion(accion, rol):
    if accion == "ver datos":
        for i, p in enumerate(personas, 1):
            print(f"{i}. {p['nombre']}")
    elif accion == "añadir persona":
        nombre = input("👤 Nombre completo: ")
        peso = float(input("⚖️ Peso (kg): "))
        altura = float(input("📏 Altura (m): "))
        personas.append({"nombre": nombre, "peso": peso, "altura": altura})
        print("✅ Persona añadida.")
    elif accion == "modificar persona":
        for i, p in enumerate(personas):
            print(f"{i + 1}. {p['nombre']}")
        idx = int(input("Seleccione persona a modificar: ")) - 1
        if 0 <= idx < len(personas):
            personas[idx]["peso"] = float(input("Nuevo peso: "))
            personas[idx]["altura"] = float(input("Nueva altura: "))
            print("✅ Datos actualizados.")
    elif accion == "eliminar persona":
        for i, p in enumerate(personas):
            print(f"{i + 1}. {p['nombre']}")
        idx = int(input("Seleccione persona a eliminar: ")) - 1
        if 0 <= idx < len(personas):
            personas.pop(idx)
            print("🗑️ Persona eliminada.")
    elif accion == "ver IMC":
        for p in personas:
            imc = round(p["peso"] / (p["altura"] ** 2), 2)
            clasif = (
                "Bajo peso" if imc < 18.5 else
                "Normal" if imc < 25 else
                "Sobrepeso" if imc < 30 else "Obesidad"
            )
            print(f"{p['nombre']}: IMC = {imc} → {clasif}")
    elif accion == "modificar roles":
        print("👮 No hay sistema de usuarios cargado en este ejemplo simple.")
    else:
        print("❌ Acción no permitida.")

# === Iniciar programa interactivo ===
print("🏋️ Bienvenido al sistema del gimnasio")
print("Selecciona tu rol:")
for r in roles:
    print(f" - {r}")

rol = input("🧑 Ingrese su rol: ").strip().lower()
if rol not in roles:
    print("⛔ Rol no válido.")
else:
    continuar = True
    while continuar:
        acciones = mostrar_menu(rol)
        opcion = input("\nEscribe el número o nombre de la acción: ").strip()
        if opcion.isdigit():
            idx = int(opcion) - 1
            if 0 <= idx < len(acciones):
                ejecutar_accion(acciones[idx], rol)
            else:
                print("❌ Opción inválida.")
        else:
            if opcion in acciones:
                ejecutar_accion(opcion, rol)
            else:
                print("❌ Acción no disponible.")
        continuar = input("\n¿Deseas realizar otra acción? (s/n): ").strip().lower() == "s"}
```

```python
import csv

# -------- Configuración de roles --------
roles = {
    "miembro": {
        "nivel": 0,
        "acciones": ["ver IMC"]
    },
    "nutricionista": {
        "nivel": 1,
        "acciones": ["ver datos", "ver IMC"]
    },
    "recepcionista": {
        "nivel": 1,
        "acciones": ["ver datos", "añadir persona"]
    },
    "entrenador": {
        "nivel": 2,
        "acciones": ["ver datos", "añadir persona", "modificar persona", "ver IMC"]
    },
    "dueno": {
        "nivel": 4,
        "acciones": ["ver datos", "añadir persona", "modificar persona", "eliminar persona", "ver IMC"]
    }
}

archivo_csv = "personas_gimnasio.csv"

# -------- Leer datos del CSV --------
def leer_personas():
    with open(archivo_csv, mode='r', newline='') as archivo:
        lector = csv.DictReader(archivo)
        return list(lector)

# -------- Guardar lista de personas --------
def guardar_personas(personas):
    campos = ["Identificación", "Nombre y Apellido", "Número Celular", "Motivo para ingresar", "Peso (kg)", "Altura (m)", "Ritmo Cardíaco (BPM)"]
    with open(archivo_csv, mode='w', newline='') as archivo:
        escritor = csv.DictWriter(archivo, fieldnames=campos)
        escritor.writeheader()
        escritor.writerows(personas)

# -------- Mostrar menú --------
def mostrar_menu(rol):
    acciones = roles[rol]["acciones"]
    print(f"\n🎮 Acciones disponibles para el rol '{rol}':")
    for i, acc in enumerate(acciones, 1):
        print(f"{i}. {acc}")
    return acciones

# -------- Ejecutar acciones --------
def ejecutar_accion(accion, rol):
    personas = leer_personas()

    if accion == "ver datos":
        for p in personas:
            print(f"{p['Nombre y Apellido']} | ID: {p['Identificación']} | Peso: {p['Peso (kg)']} kg | Altura: {p['Altura (m)']} m")

    elif accion == "ver IMC":
        for p in personas:
            peso = float(p["Peso (kg)"])
            altura = float(p["Altura (m)"])
            imc = round(peso / (altura ** 2), 2)
            clasif = (
                "Bajo peso" if imc < 18.5 else
                "Normal" if imc < 25 else
                "Sobrepeso" if imc < 30 else "Obesidad"
            )
            print(f"{p['Nombre y Apellido']}: IMC = {imc} → {clasif}")

    elif accion == "añadir persona":
        nueva = {
            "Identificación": input("ID: "),
            "Nombre y Apellido": input("Nombre completo: "),
            "Número Celular": input("Celular: "),
            "Motivo para ingresar": input("Motivo: "),
            "Peso (kg)": input("Peso (kg): "),
            "Altura (m)": input("Altura (m): "),
            "Ritmo Cardíaco (BPM)": input("BPM: ")
        }
        personas.append(nueva)
        guardar_personas(personas)
        print("✅ Persona añadida.")

    elif accion == "modificar persona":
        for i, p in enumerate(personas):
            print(f"{i+1}. {p['Nombre y Apellido']}")
        idx = int(input("Selecciona el número de persona a modificar: ")) - 1
        if 0 <= idx < len(personas):
            personas[idx]["Peso (kg)"] = input("Nuevo peso: ")
            personas[idx]["Altura (m)"] = input("Nueva altura: ")
            guardar_personas(personas)
            print("✅ Datos actualizados.")

    elif accion == "eliminar persona":
        for i, p in enumerate(personas):
            print(f"{i+1}. {p['Nombre y Apellido']}")
        idx = int(input("Selecciona el número de persona a eliminar: ")) - 1
        if 0 <= idx < len(personas):
            personas.pop(idx)
            guardar_personas(personas)
            print("🗑️ Persona eliminada.")

# -------- Iniciar sistema --------
print("🏋️ Bienvenido al sistema del gimnasio")
print("Selecciona tu rol:")
for r in roles:
    print(f" - {r}")

rol = input("🧑 Ingresa tu rol: ").strip().lower()
if rol not in roles:
    print("❌ Rol no válido.")
else:
    continuar = True
    while continuar:
        acciones = mostrar_menu(rol)
        opcion = input("\nEscribe el número o nombre de la acción: ").strip()
        if opcion.isdigit():
            idx = int(opcion) - 1
            if 0 <= idx < len(acciones):
                ejecutar_accion(acciones[idx], rol)
            else:
                print("❌ Opción inválida.")
        elif opcion in acciones:
            ejecutar_accion(opcion, rol)
        else:
            print("❌ Acción no permitida.")
        continuar = input("\n¿Deseas realizar otra acción? (s/n): ").lower() == "s"
```
