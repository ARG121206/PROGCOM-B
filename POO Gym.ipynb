```python
import csv

# -------- Configuraci√≥n de roles --------
roles = {
    "miembro": {
        "nivel": 0,
        "acciones": ["ver IMC"]
    },
    "nutricionista": {
        "nivel": 1,
        "acciones": ["ver datos", "ver IMC"]
    },
    "recepcionista": {
        "nivel": 1,
        "acciones": ["ver datos", "a√±adir persona"]
    },
    "entrenador": {
        "nivel": 2,
        "acciones": ["ver datos", "a√±adir persona", "modificar persona", "ver IMC"]
    },
    "dueno": {
        "nivel": 4,
        "acciones": ["ver datos", "a√±adir persona", "modificar persona", "eliminar persona", "ver IMC"]
    }
}

archivo_csv = "personas_gimnasio.csv"

# -------- Leer datos del CSV --------
def leer_personas():
    with open(archivo_csv, mode='r', newline='') as archivo:
        lector = csv.DictReader(archivo)
        return list(lector)

# -------- Guardar lista de personas --------
def guardar_personas(personas):
    campos = ["Identificaci√≥n", "Nombre y Apellido", "N√∫mero Celular", "Motivo para ingresar", "Peso (kg)", "Altura (m)", "Ritmo Card√≠aco (BPM)"]
    with open(archivo_csv, mode='w', newline='') as archivo:
        escritor = csv.DictWriter(archivo, fieldnames=campos)
        escritor.writeheader()
        escritor.writerows(personas)

# -------- Mostrar men√∫ --------
def mostrar_menu(rol):
    acciones = roles[rol]["acciones"]
    print(f"\nüéÆ Acciones disponibles para el rol '{rol}':")
    for i, acc in enumerate(acciones, 1):
        print(f"{i}. {acc}")
    return acciones

# -------- Ejecutar acciones --------
def ejecutar_accion(accion, rol):
    personas = leer_personas()

    if accion == "ver datos":
        for p in personas:
            print(f"{p['Nombre y Apellido']} | ID: {p['Identificaci√≥n']} | Peso: {p['Peso (kg)']} kg | Altura: {p['Altura (m)']} m")

    elif accion == "ver IMC":
        for p in personas:
            peso = float(p["Peso (kg)"])
            altura = float(p["Altura (m)"])
            imc = round(peso / (altura ** 2), 2)
            clasif = (
                "Bajo peso" if imc < 18.5 else
                "Normal" if imc < 25 else
                "Sobrepeso" if imc < 30 else "Obesidad"
            )
            print(f"{p['Nombre y Apellido']}: IMC = {imc} ‚Üí {clasif}")

    elif accion == "a√±adir persona":
        nueva = {
            "Identificaci√≥n": input("ID: "),
            "Nombre y Apellido": input("Nombre completo: "),
            "N√∫mero Celular": input("Celular: "),
            "Motivo para ingresar": input("Motivo: "),
            "Peso (kg)": input("Peso (kg): "),
            "Altura (m)": input("Altura (m): "),
            "Ritmo Card√≠aco (BPM)": input("BPM: ")
        }
        personas.append(nueva)
        guardar_personas(personas)
        print("‚úÖ Persona a√±adida.")

    elif accion == "modificar persona":
        for i, p in enumerate(personas):
            print(f"{i+1}. {p['Nombre y Apellido']}")
        idx = int(input("Selecciona el n√∫mero de persona a modificar: ")) - 1
        if 0 <= idx < len(personas):
            personas[idx]["Peso (kg)"] = input("Nuevo peso: ")
            personas[idx]["Altura (m)"] = input("Nueva altura: ")
            guardar_personas(personas)
            print("‚úÖ Datos actualizados.")

    elif accion == "eliminar persona":
        for i, p in enumerate(personas):
            print(f"{i+1}. {p['Nombre y Apellido']}")
        idx = int(input("Selecciona el n√∫mero de persona a eliminar: ")) - 1
        if 0 <= idx < len(personas):
            personas.pop(idx)
            guardar_personas(personas)
            print("üóëÔ∏è Persona eliminada.")

# -------- Iniciar sistema --------
print("üèãÔ∏è Bienvenido al sistema del gimnasio")
print("Selecciona tu rol:")
for r in roles:
    print(f" - {r}")

rol = input("üßë Ingresa tu rol: ").strip().lower()
if rol not in roles:
    print("‚ùå Rol no v√°lido.")
else:
    continuar = True
    while continuar:
        acciones = mostrar_menu(rol)
        opcion = input("\nEscribe el n√∫mero o nombre de la acci√≥n: ").strip()
        if opcion.isdigit():
            idx = int(opcion) - 1
            if 0 <= idx < len(acciones):
                ejecutar_accion(acciones[idx], rol)
            else:
                print("‚ùå Opci√≥n inv√°lida.")
        elif opcion in acciones:
            ejecutar_accion(opcion, rol)
        else:
            print("‚ùå Acci√≥n no permitida.")
        continuar = input("\n¬øDeseas realizar otra acci√≥n? (s/n): ").lower() == "s"
```
