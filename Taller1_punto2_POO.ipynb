```python
class Paciente:
    def __init__(self, nombre, saturacion_oxigeno, cantidad_vacunas, prueba_covid):
        self.nombre = nombre
        self.saturacion_oxigeno = saturacion_oxigeno
        self.cantidad_vacunas = cantidad_vacunas
        self.prueba_covid = prueba_covid

class SistemaTriage:
    def __init__(self, capacidad_urgencias=10):
        self.camillas_disponibles = capacidad_urgencias

    def evaluar_paciente(self, paciente):
        print(f"\nEvaluando a {paciente.nombre}...")

        if paciente.saturacion_oxigeno < 90:
            print("→ Sala COVID (saturación baja)")
        elif paciente.cantidad_vacunas < 2:
            print("→ Sala COVID (pocas vacunas)")
        elif paciente.prueba_covid:
            print("→ Sala COVID (prueba positiva)")
        else:
            if self.camillas_disponibles > 0:
                self.camillas_disponibles -= 1
                print("→ Urgencias (camilla asignada)")
            else:
                print("→ Espera (no hay camillas disponibles)")

```
